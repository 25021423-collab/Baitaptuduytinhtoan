```python
h = int(input("Nhập độ cao h: "))

for i in range(1, h + 1):
    # 1. In khoảng trắng đệm bên trái để tạo độ nghiêng cho tam giác
    print(" " * (h - i), end="")
    
    # 2. Xử lý phần thân tam giác (vùng chứa dấu sao và lỗ hổng)
    for j in range(1, 2 * i):
        # Điều kiện in dấu sao:
        # j == 1: cạnh trái
        # j == 2*i - 1: cạnh phải
        # i == h: cạnh đáy
        if j == 1 or j == 2 * i - 1 or i == h:
            print("*", end="")
        else:
            print(" ", end="") # In khoảng trống ở giữa
            
    # Xuống dòng sau mỗi tầng
    print()
```

```python
max_val = None
min_val = None
while True:
    x = int(input())
    if x == -1:
        break
    if max_val is None or x > max_val:
        max_val = x
    if min_val is None or x < min_val:
        min_val = x
print("Lớn nhất:", max_val)
print("Nhỏ nhất:", min_val)
```

```python

n = int(input())
even_count = odd_count = 0
temp = n
while temp > 0:
    digit = temp % 10
    if digit % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
    temp //= 10
print("Chữ số chẵn:", even_count)
print("Chữ số lẻ:", odd_count)
```

```python
def is_prime(x):
    if x < 2:
        return False
    i = 2
    while i * i <= x:
        if x % i == 0:
            return False
        i += 1
    return True

n = int(input())
count = 0
temp = n
while temp > 0:
    digit = temp % 10
    if is_prime(digit):
        count += 1
    temp //= 10
print(count)
```
