#### Bài 6:
"[SumPrimesInRange]
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."


```python
import math
import sys

def la_so_nguyen_to(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True
print("Nhập các bộ test (a b): ")

for line in sys.stdin:
    try:
        data = list(map(int, line.split()))
        if len(data) < 2:
            continue
        
        a, b = data[0], data[1]
        tong_snt = 0
       
        for x in range(a, b + 1):
            if la_so_nguyen_to(x):
                tong_snt += x
        
        print(f"Tổng các số nguyên tố trong [{a}, {b}] là: {tong_snt}")
        
    except ValueError:
        print("Vui lòng nhập hai số nguyên hợp lệ.")
```

```python
def is_prime(n):
    """Check if a number is prime"""
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

def sum_primes_in_range(a, b):
    """Calculate sum of prime numbers in range [a, b]"""
    total = 0
    for num in range(a, b + 1):
        if is_prime(num):
            total += num
    return total

# Read input and process test cases
try:
    while True:
        line = input().strip()
        if not line:
            break
        a, b = map(int, line.split())
        result = sum_primes_in_range(a, b)
        print(result)
except EOFError:
    pass
```

#### Bài 8:
ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 


```python
n = int(input())
def reverse(num):
    rev = 0
    while num > 0:
        rev = rev * 10 + (num % 10)
        num //= 10
    return rev
def palindrome(num):
    return num == reverse(num)

buoc = 0

while not palindrome(n):
    n = n + reverse(n)
    buoc += 1
print(buoc, n)
```

#### Bài 10:
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 


```python

n = int(input())
def get_collatz_length(x):
    length = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3 * x + 1
        length += 1
    return length
max_length = 0
best_x = 1

for i in range(1, n + 1):
    current_length = get_collatz_length(i)
    
    if current_length > max_length:
        max_length = current_length
        best_x = i

print(best_x, max_length)
```

```python
def collatz_length(n):
    length = 1
    while n != 1:
        n = n // 2 if n % 2 == 0 else 3 * n + 1
        length += 1
    return length

n = int(input())

best_x, best_len = 1, 1
for x in range(1, n + 1):
    l = collatz_length(x)
    if l > best_len:
        best_x, best_len = x, l

print(best_x, best_len)
```

#### Bài 13:
ClosePairNumber]
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false" 


```python
m, n = map(int, input().split())
def tong(num):
    total = 0
 
    for i in range(1, (num // 2) + 1):
        if num % i == 0:
            total += i
    return total

if tong(m) == n and tong(n) == m:
    print("true")
else:
    print("false")
```
